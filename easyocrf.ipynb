```python
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('sign2.jpg')

plt.imshow(img);
```

```python
reader=easyocr.Reader(['en','mr'])
```

```python
result=reader.readtext('sign2.jpg')
result
```

```python
result[3][1]
```

```python
len(result)
```

```python
for i in range(len(result)):
    print(result[i][1])
```

```python
with open('new.txt','w') as f:
    for i in range(len(result)):
        f.write(str(result[i][1]) + '\n')
```

```python
result[0][0]
```

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
```

```python
for i in range(len(result)):
    img=cv2.putText(img,result[i][1],top_left,cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)
    img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),2)

```

```python
img=cv2.imread('sign2.jpg')
result=reader.readtext(img)
result
spacer=100
for i in result:
    top_left=tuple(i[0][0])
    bottom_right=tuple(i[0][2])
    text=i[1]
    img=cv2.putText(img,text,top_left,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
    img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),2)
    spacer+=15


plt.imshow(img);
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
